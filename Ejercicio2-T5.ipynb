```python
import math
from math import log
import matplotlib.pyplot as plt
```

```python
def factorial(n):
   if n==0 or n==1:
            resultado=1
   elif n>1:
            resultado=n*factorial(n-1)
   return resultado
```

```python
lista1=[]
lista2=[]
for i in range(0,21): 
    K=1.380649*(10**23)
    x=299+i
    fac1=factorial(x)
    fac2=factorial(i)
    total=(fac1)/(factorial(299)*fac2)
    lista1.append(total)
    entropia=K*log(total)
    lista2.append(entropia)
    
    
```

```python
lista3=[]
lista4=[]
listaaa=[]

num=100
for i in range(0,num+1):
    q=num-i
    listaaa.append(q)

for i in listaaa:
    K=1.380649*(10**23)
    x=199+i
    fac1=factorial(x)
    fac2=factorial(i)
    total=(fac1)/(factorial(199)*fac2)
    lista3.append(total)
    entropia=K*log(total)
    lista4.append(entropia)
    
```

```python
stotal=[]
for i in range(len(lista2)):
	stotal.append(lista2[i]+lista4[i])
```

```python
def numero_mayor (lista):
    max_valor=None 
    for i in lista:
        if (max_valor is None or i>max_valor): 
            max_valor=i
            posicion=lista.index(i) 
            
    print('El valor máximo es', max_valor , 'y se encuentra en la posición', posicion)
```

```python
numero_mayor(stotal) #Esto para ver en dónde se ubica el valor máximo de la entropía total
```

```python
#Graficando
plt.plot(stotal)
plt.plot(lista2)
plt.plot(lista4)
plt.title('Entropías')
plt.plot(60,3.6507304040009273e+25,marker ="o")
plt.show()


```
