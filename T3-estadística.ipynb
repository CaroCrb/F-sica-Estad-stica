```python
#Problema 1

#Zn es la suma de n términos obtenidos de la distribución uniforme U

import numpy as np
from numpy import random as r
from random import choices
import matplotlib.pyplot as plt
from scipy.stats import uniform
from scipy.stats import norm
import statistics
```

```python
#Definimos la función Zn

def Zn(N):
    lista=[]
    for i in range (N):
        x=uniform.rvs(-1,5,size=100) #Función para la distribución uniforme (centrada en 2) y con N=100 términos
        lista.append(sum(x)) #Enviamos los valores a la lista 'lista'
    return lista
    
```

```python
#Con este código sabremos la frecuencia de aparición de cada valor
sumas=Zn(10000)
numeros={}
for elemento in lista:
    if elemento in numeros:
        numeros[elemento]+=1
    else:
        numeros[elemento]=1

for valor in sorted(numeros):
    print(f'{valor}: {numeros[valor]}')
```

```python
#Ahora hacemos el histograma 

plt.hist(sumas,bins=25,color='darkcyan', rwidth=0.8)
plt.title("Histograma de los valores Zn generados")
plt.xlabel("Total")
plt.ylabel("Frecuencia")
```

```python
#Haciendo ahora la distribución Gaussiana 

media=statistics.mean(sumas)
desviacion=np.var(sumas)**0.5
normal=np.arange(100,200,0.3)
grafica=plt.plot(normal,norm.pdf(normal,media,desviacion),color='purple')
plt.title("Ajuste Gaussiano de los datos obtenidos")
plt.xlabel("Total")
plt.ylabel("Probabilidad")

#Para graficar busqué sobre las funciones estadísticas que tiene python en https://docs.scipy.org/doc/scipy/reference/stats.html
```

```python
#Problema 2

#Definimos primera la nueva función 
def VA(N):
    lista2=[]
    for i in range(N):
        X_1=norm.rvs(0,1,size=10000) #Distribucion normal N(0,1)=X_1
        X_2=norm.rvs(-1,1,size=10000) #Distribucion normal N(-1,1)=X_2
        lista2.append(X_1+X_2) #Como la nueva variable Y es la suma de las variables anteriores,
                               # entonces la añadimos a la nueva lista 'lista2'
        print(lista2) #Impresión de la lista con todos los valores de Y
    return lista2
```

```python
z=VA(1)
```

```python
#Elaborando el histograma 

plt.hist(z, bins=30, color='darkcyan', rwidth=0.8)
plt.xlabel("Nueva variable Y")
plt.ylabel("Frecuencia")
plt.title("Histograma para los valores obtenidos de Y")
```

```python
#Haciendo ahora la distribución Gaussiana

media2=0  #media
desv=2**0.5 #varianza
normal2=np.arange(-6,6,0.01)
plt.plot(normal2,norm.pdf(normal2,media2,desv),color='purple')
plt.title("Ajuste Gaussiano de los datos obtenidos")
plt.xlabel("Total")
plt.ylabel("Probabilidad")
```

```python
#Problema 3

def VA2(N):
    lista3=[]
    for i in range(N):
        X_3=norm.rvs(0,1,size=10000)
        lista3.append(4*X_3+6)
    print(lista3)
    return lista3

#Notemos que esta función es muy parecida a la del problema anterior
```

```python
k=VA2(1) #Se imprimen todas los nuevos valores de la variable Y
```

```python
#Elaborando el histograma 

plt.hist(k, bins=30, color='darkcyan', rwidth=0.8)
plt.xlabel("Nueva variable Y")
plt.ylabel("Frecuencia")
plt.title("Histograma para los valores generados de Y")
```

```python
#Haciendo ahora la distribución Gaussiana

media3=6  #media
desv3=(np.var(k)*(16))**0.5 #varianza
normal3=np.arange(-50,50,0.01)
plt.plot(normal3,norm.pdf(normal3,media3,desv3),color='purple')
plt.title("Ajuste Gaussiano de los datos obtenidos")
plt.xlabel("Total")
plt.ylabel("Probabilidad")
```
