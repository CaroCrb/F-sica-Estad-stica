```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import math as math
```

```python
T1=6
T2=7
T3=8
T4=9
T5=10
```

```python
#Declarando las funciones F con cada valor de T respectivamente

def F1 (x):
    return (-4)*(x**2)+T1*(((1+x)/2)*log((1+x)/2)+(((1-x)/2)*log((1-x)/2)))

def F2 (x):
    return (-4)*(x**2)+T2*(((1+x)/2)*log((1+x)/2)+(((1-x)/2)*log((1-x)/2)))

def F3 (x):
    return (-4)*(x**2)+T3*(((1+x)/2)*log((1+x)/2)+(((1-x)/2)*log((1-x)/2)))

def F4 (x):
    return (-4)*(x**2)+T4*(((1+x)/2)*log((1+x)/2)+(((1-x)/2)*log((1-x)/2)))


def F5 (x):
    return (-4)*(x**2)+T5*(((1+x)/2)*log((1+x)/2)+(((1-x)/2)*log((1-x)/2)))

x = np.arange(-.99, .99, 0.01) 

#Graficando
plt.plot(x, [F1(i) for i in x], "grey")
plt.plot(x, [F2(i) for i in x], "k")
plt.plot(x, [F3(i) for i in x], "r")
plt.plot(x, [F4(i) for i in x], "orange")
plt.plot(x, [F5(i) for i in x], "g")
plt.ylabel('F*')
plt.xlabel('-1<x<1') 

plt.show()
```
