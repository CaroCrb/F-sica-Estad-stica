```python
import numpy as np
from numpy import random as r
import matplotlib.pyplot as plt
```

```python
def caminante(N):
    pasos=r.choice([-1,0,1],N) #Elegimos el -1,0,1 como los valores que pueden tomar los pasos
    caminante=np.cumsum(pasos) #Realizamos la suma de los pasos
    
    return(caminante)
```

```python
def f1(x):
    return (2*x/3)**(1/2)
def f2(x):
    return -(2*x/3)**(1/2)
x=range(0,30)
```
